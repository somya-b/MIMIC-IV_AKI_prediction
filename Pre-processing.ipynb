```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/SHP6004/sph6004_assignment1_data.csv')

df_copy = df.copy()
```

```python
df.head()
```

```python
df.shape
```

```python
df.loc[df.aki > 0, 'aki'] = 1
df.head()
```

```python
df.describe().transpose()
```

```python
#replacing na lab result values with pcot values
pcot = df.columns[df_copy.columns.str.endswith('.1')]

for col in pcot:
  lab = col.replace('.1','')
  df[lab].fillna(df[col], inplace=True)

#removing pcot columns
df.drop(pcot, axis=1, inplace=True)

df['glucose_min'].fillna(df['glucose_min.2'], inplace=True)
df['glucose_max'].fillna(df['glucose_max.2'], inplace=True)
df.drop(['glucose_min.2','glucose_max.2'], axis=1, inplace=True)

df['aado2_min'].fillna(df['aado2_calc_min'],inplace=True)
df['aado2_max'].fillna(df['aado2_calc_max'],inplace=True)
df.drop(['aado2_calc_max','aado2_calc_min'], axis=1, inplace=True)
```

```python
df.shape
```

```python
# remove rows with more than 50% of missing columns
df = df.dropna(axis=0, thresh=int(0.5 * df.shape[1]))

#removing columns with more than 20% data missing
df = df.dropna(axis=1, thresh=int(0.2*df.shape[0]))

df.shape
```

```python
import plotly.graph_objects as go
import plotly.express as px

gcs_df = pd.DataFrame(
    {
        'Total GCS (min)':np.hstack([df.gcs_min,df.gcs_min,df.gcs_min,df.gcs_min]),
        'Scale':np.hstack([df.gcs_verbal,df.gcs_motor,df.gcs_eyes, df.gcs_unable]),
        'Type':['Verbal']*len(df.gcs_verbal)+['Motor']*len(df.gcs_motor)+['Eyes']*len(df.gcs_eyes)+['Unable to record']*len(df.gcs_unable)
    }
)

fig = px.scatter(gcs_df,x='Scale',y='Total GCS (min)',facet_col='Type',color='Type')
fig
```

```python
#to see if the relation is not because of a singular point
print(len(df.loc[df.gcs_unable==1]))
print(len(df.loc[df.gcs_verbal==0]))

df.gcs_min[df.gcs_verbal==0].fillna(15.0, inplace=True)
df.gcs_min[df.gcs_unable==1].fillna(15.0, inplace=True)
```

```python
i = df.loc[df['gcs_min'].isnull()].index
j = df.loc[df[['gcs_verbal','gcs_motor','gcs_eyes']].isnull().any(axis=1)].index

print(len(i))
print(i.isin(j).sum())
```

```python
cols = df.select_dtypes(include='float64').columns

df[cols] = (df[cols]-df[cols].mean())/df[cols].std()
```

```python
#MICE imputaion (https://medium.com/@brijesh_soni/topic-9-mice-or-multivariate-imputation-with-chain-equation-f8fd435ca91)
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

missing_mask = df.iloc[:,5:].isna()
imputer = IterativeImputer(max_iter=100, random_state=0)
df.iloc[:,5:] = imputer.fit_transform(df.iloc[:,5:])

df.isnull().sum().sum()
```

```python
df.shape
```

```python
from scipy import stats

numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

# First, calculate the z-score for each row.
z_scores = np.abs(stats.zscore(df[numeric_columns]))

# Then, select the rows where the z-score is greater than 3 for any column.
outlier_rows = np.where(z_scores > 3)[0]
outlier_rows = pd.Series(outlier_rows).unique()
outlier_rows = df.index[outlier_rows]

# Finally, remove the outlier rows from the DataFrame.
df = df.drop(outlier_rows, axis=0)

df.shape
```

```python
df.drop(['gcs_verbal','gcs_motor','gcs_eyes','gcs_unable'],axis=1,inplace=True)
```

```python
from scipy import stats

#removing variables which do not vary with group
aki_0 = df.loc[df.aki==0,:]
aki_1 = df.loc[df.aki>0,:]

numeric_columns = df.select_dtypes(include=['float64']).columns

t_test = pd.DataFrame({'variables':numeric_columns, 't_val':0, 'p_val':0})

i=0
for col in numeric_columns:
  t,p = stats.ttest_ind(aki_0[col], aki_1[col])
  t_test.iloc[i,1:3] = [t,p]
  i=i+1

t_test = t_test.iloc[2:,:] #removing id & aki
vars = t_test.loc[t_test.p_val>.05,'variables'] #variables which can't reject the null of two sample t-test
print(vars)

df.drop(vars, axis=1, inplace=True)
```

```python
confusion_matrix = pd.crosstab(df['aki'], df['gender'])
print(stats.contingency.association(confusion_matrix, method="cramer"))

confusion_matrix = pd.crosstab(df['aki'], df['race'])
print(stats.contingency.association(confusion_matrix, method="cramer"))
```

```python
#transforming categorical variables
race = pd.get_dummies(df.race)

print(race.shape)

race['ASIAN'] = race.loc[:,'ASIAN':'ASIAN - SOUTH EAST ASIAN'].sum(axis=1)
race.drop(race.loc[:,'ASIAN - ASIAN INDIAN':'ASIAN - SOUTH EAST ASIAN'].columns, axis=1, inplace=True)

race['BLACK'] = race.loc[:,'BLACK/AFRICAN':'BLACK/CARIBBEAN ISLAND'].sum(axis=1)
race.drop(race.loc[:,'BLACK/AFRICAN':'BLACK/CARIBBEAN ISLAND'].columns, axis=1, inplace=True)

race['HISPANIC'] = race.loc[:,'HISPANIC OR LATINO':'HISPANIC/LATINO - SALVADORAN'].sum(axis=1)
race.drop(race.loc[:,'HISPANIC OR LATINO':'HISPANIC/LATINO - SALVADORAN'].columns, axis=1, inplace=True)

race['WHITE'] = race.loc[:,'WHITE':'WHITE - RUSSIAN'].sum(axis=1)
race.drop(race.loc[:,'WHITE - BRAZILIAN':'WHITE - RUSSIAN'].columns, axis=1, inplace=True)

race['OTHER'] = race.loc[:,['OTHER','AMERICAN INDIAN/ALASKA NATIVE','MULTIPLE RACE/ETHNICITY','NATIVE HAWAIIAN OR OTHER PACIFIC ISLANDER','PORTUGUESE','SOUTH AMERICAN']].sum(axis=1)
race.drop(race.loc[:,['AMERICAN INDIAN/ALASKA NATIVE','MULTIPLE RACE/ETHNICITY','NATIVE HAWAIIAN OR OTHER PACIFIC ISLANDER','PORTUGUESE','SOUTH AMERICAN']].columns, axis=1, inplace=True)

race['UNKNOWN'] = race.loc[:,['UNKNOWN','PATIENT DECLINED TO ANSWER','UNABLE TO OBTAIN']].sum(axis=1)
race.drop(race.loc[:,['PATIENT DECLINED TO ANSWER','UNABLE TO OBTAIN']].columns, axis=1, inplace=True)

print(race.shape)

#to check if the transformation was correct
for column in race.columns:
    print(f"{column}: {race[column].nunique()}")

print(race.sum())
```

```python
df.drop('race',axis=1,inplace=True)

df = df.join(race)
```
